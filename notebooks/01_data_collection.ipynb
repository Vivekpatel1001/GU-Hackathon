# Data Collection
Load Instagram, YouTube, and Twitter CSV datasets from the project data folder.

```python
from pathlib import Path
import pandas as pd

base = Path('..') / 'data'
instagram = pd.read_csv(base / 'Instagram_Analytics.csv')
youtube = pd.read_csv(base / 'youtube_trending_data.csv')
twitter = pd.read_csv(base / 'training.1600000.processed.noemoticon (1).csv', encoding='latin-1', header=None, nrows=50000)

print('Instagram', instagram.shape)
print('YouTube', youtube.shape)
print('Twitter', twitter.shape)
```
