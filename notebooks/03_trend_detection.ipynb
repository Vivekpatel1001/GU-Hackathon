# Trend Detection
Frequency + TF-IDF + simple spike detection for social trend signals.

```python
import re
from collections import Counter
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from pathlib import Path

yt = pd.read_csv(Path('..') / 'data' / 'youtube_trending_data.csv')
titles = yt['Title'].dropna().astype(str).tolist() if 'Title' in yt.columns else []

vectorizer = TfidfVectorizer(max_features=20, stop_words='english')
matrix = vectorizer.fit_transform(titles) if titles else None
keywords = vectorizer.get_feature_names_out().tolist() if titles else []
print('Top keywords:', keywords[:10])

hashtags = []
for t in titles[:1000]:
    hashtags.extend(re.findall(r'#(\w+)', t.lower()))
print('Top hashtags:', Counter(hashtags).most_common(10))
```
