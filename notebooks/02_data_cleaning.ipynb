# Data Cleaning
Normalize text and handle missing values for all input sources.

```python
import re
import pandas as pd
from pathlib import Path

base = Path('..') / 'data'
ig = pd.read_csv(base / 'Instagram_Analytics.csv')

def clean_text(text):
    text = '' if pd.isna(text) else str(text)
    text = re.sub(r'http\S+|www\.\S+', '', text)
    text = re.sub(r'[^\w\s#@.,!?-]', '', text)
    return re.sub(r'\s+', ' ', text).strip()

if 'caption' in ig.columns:
    ig['caption_clean'] = ig['caption'].apply(clean_text)

for col in ['likes', 'comments', 'shares', 'saves', 'reach']:
    if col in ig.columns:
        ig[col] = pd.to_numeric(ig[col], errors='coerce').fillna(0)

clean_dir = base / 'cleaned'
clean_dir.mkdir(exist_ok=True)
ig.to_csv(clean_dir / 'instagram_cleaned.csv', index=False)
print('Saved cleaned Instagram data')
```
