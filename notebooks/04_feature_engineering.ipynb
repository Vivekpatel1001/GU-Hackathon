# Feature Engineering
Prepare model-ready virality features from Instagram analytics.

```python
import pandas as pd
from pathlib import Path

df = pd.read_csv(Path('..') / 'data' / 'Instagram_Analytics.csv')
for col in ['likes', 'comments', 'shares', 'saves', 'reach']:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

df['total_engagement'] = df['likes'] + df['comments'] + df['shares'] + df['saves']
df['engagement_per_reach'] = (df['total_engagement'] / df['reach'].replace(0, 1)) * 100
df['comment_to_like_ratio'] = df['comments'] / df['likes'].replace(0, 1)
df['save_rate'] = (df['saves'] / df['reach'].replace(0, 1)) * 100

print(df[['total_engagement', 'engagement_per_reach', 'comment_to_like_ratio', 'save_rate']].head())
```
